```python
# Check resources... not super conclusive.
# The resource ratio is the number of unique resources that are not present in all plans over the number of unique resources present across all plans.

import json

from app.utils.action_plan_types import ActionPlan


def calculate_max_resource_dist(client_id):
    # Get all rows for that client, and extract structured plan and markdown
    gen_data = base[base["inputs.client_id"] == client_id]["outputs.gen_data"]
    plans = [
        ActionPlan(**json.loads(data)["structured_action_plan"]) for data in gen_data
    ]
    max_dist = float("-inf")
    for i in range(len(plans)):
        for j in range(i + 1, len(plans)):  # start from i+1 to avoid duplicates
            dist = calculate_resource_ratio(plans[i], plans[j])
            max_dist = max(max_dist, dist)
    return max_dist


def calculate_resource_ratio(plan_one, plan_two):
    all_resources = {}
    r_ids = []
    for plan_index, plan in enumerate([plan_one, plan_two]):
        for section in plan.sections:
            for resource in section.resources:
                if resource.id in r_ids:
                    all_resources[resource.id]["count"] += 1
                    all_resources[resource.id]["sections"].append(section.title)
                    if plan_index not in all_resources[resource.id]["plan_indexes"]:
                        all_resources[resource.id]["plan_indexes"].append(plan_index)

                else:
                    all_resources[resource.id] = {
                        "count": 1,
                        "resource": resource,
                        "sections": [section.title],
                        "plan_indexes": [plan_index],
                    }
                    r_ids.append(resource.id)

    resource_diff = {
        k: v for k, v in all_resources.items() if len(v["plan_indexes"]) != 2
    }
    return len(resource_diff) / len(all_resources)


base_diff["resource_ratio"] = base_diff.index.map(calculate_max_resource_dist)
```

```python
# Compute content difference with embeddings
from functools import cache

from openai import OpenAI

from app.core.config import settings

client = OpenAI(api_key=settings.OPENAI_API_KEY)


# Cache the results so we don't fetch the same embeddings twice
@cache
def get_embeddings(text: str):
    response = client.embeddings.create(input=text, model="text-embedding-3-small")
    data = response.data[0]
    return data.embedding
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


# Function that takes the sections and title embeddings of two plans and matches the titles in pairs, leaving out unconvincing matches.
def match_titles_with_embeddings(sections_one, sections_two):
    """
    Match titles from two lists based on semantic similarity.
    Args:
        sections_one: List of dicts, each with {'title_embeddings': np.array, 'section': ActionPlanSection}
        sections_two: List of dicts, each with {'title_embeddings': np.array, 'section': ActionPlanSection}
    Returns:
        List of dicts containing matched sections and their similarity scores
    """
    # Extract embeddings and convert to numpy arrays
    embeddings_one = np.array([item["title_embeddings"] for item in sections_one])
    embeddings_two = np.array([item["title_embeddings"] for item in sections_two])

    # Calculate cosine similarity between all pairs
    similarity_matrix = cosine_similarity(embeddings_one, embeddings_two)

    # Create initial pairs with all similarity scores
    all_pairs = []
    for i, section_one in enumerate(sections_one):
        for j, section_two in enumerate(sections_two):
            all_pairs.append(
                (
                    section_one["section"],
                    section_two["section"],
                    similarity_matrix[i][j],
                    i,
                    j,
                )
            )

    # Sort by similarity score
    all_pairs.sort(key=lambda x: x[2], reverse=True)

    # Match pairs greedily, ensuring each title is only used once
    matched_ones = set()
    matched_twos = set()
    final_pairs = []

    for section_one, section_two, score, i, j in all_pairs:
        if i not in matched_ones and j not in matched_twos:
            final_pairs.append(
                {
                    "section_one": [section_one],
                    "section_two": [section_two],
                    "title_similarity": score,
                    "i": i,
                    "j": j,
                }
            )
            matched_ones.add(i)
            matched_twos.add(j)

        if len(final_pairs) == min(len(sections_one), len(sections_two)):
            break

    # Handle low similarity scores
    bad_pairs = []
    good_pairs = []

    for pair in final_pairs:
        # 0.4 was found experimentally
        # running the eval prints the good and bad pairs and I found this number
        # correctly paired remotely related sections while avoiding totally unrelated pairings
        if pair["title_similarity"] < 0.4:
            bad_pairs.append(pair)
            print(
                [s.title for s in pair["section_one"]],
                [s.title for s in pair["section_two"]],
            )
        else:
            good_pairs.append(pair)

    unmatched_one = []
    unmatched_two = []

    # For each bad pair, try to find a better match
    for pair in bad_pairs:
        # Find best similarity score for section_one
        best_sim_one = max(similarity_matrix[pair["i"]])
        if best_sim_one > 0.4:
            best_j_one = np.argmax(similarity_matrix[pair["i"]])
            for good_pair in good_pairs:
                if good_pair["j"] == best_j_one:
                    good_pair["section_one"].append(pair["section_one"][0])
                    break
        else:
            unmatched_one.append(pair["section_one"][0])

        # Find best similarity score for section_two
        best_sim_two = max(similarity_matrix[:, pair["j"]])
        if best_sim_two > 0.4:
            best_i_two = np.argmax(similarity_matrix[:, pair["j"]])
            for good_pair in good_pairs:
                if good_pair["i"] == best_i_two:
                    good_pair["section_two"].append(pair["section_two"][0])
                    break
        else:
            unmatched_two.append(pair["section_two"][0])

    return (good_pairs, unmatched_one, unmatched_two)
```

```python
# simply computes and saves the content embeddings for each group of sections
# computes and saves the dot products between the groups
def compare_matched(matched):
    similarities_per_sections = []
    for match in matched:
        p1_s_content = ""
        p2_s_content = ""

        for section in match["section_one"]:
            p1_s_content += section.markdown_content
            p1_s_content += "/n"

        for section in match["section_two"]:
            p2_s_content += section.markdown_content
            p1_s_content += "/n"

        section1_embeddings = get_embeddings(p1_s_content)
        section2_embeddings = get_embeddings(p2_s_content)
        similarities_per_sections.append(
            {
                "section_one": match["section_one"],
                "section_two": match["section_two"],
                "title_similarity": match["title_similarity"],
                "content_similarity": np.dot(section1_embeddings, section2_embeddings),
                "content_embeddings_one": section1_embeddings,
                "content_embeddings_two": section2_embeddings,
            }
        )

    return similarities_per_sections
```

```python
# pairs the titles
# compute the minimum content similarity for paired sections
# also return the ratio of all unmatched sections over the total amount of sections
# and the titles of the unmatched sections
def get_content_similarity(plan_one, plan_two):
    plan_one_title_embeddings = [
        {"section": section, "title_embeddings": get_embeddings(section.title)}
        for section in plan_one.sections
    ]

    plan_two_title_embeddings = [
        {"section": section, "title_embeddings": get_embeddings(section.title)}
        for section in plan_two.sections
    ]

    good_pairs, unmatched_one, unmatched_two = match_titles_with_embeddings(
        plan_one_title_embeddings, plan_two_title_embeddings
    )
    print(
        [
            [
                [s.title for s in pair["section_one"]],
                [s.title for s in pair["section_two"]],
            ]
            for pair in good_pairs
        ]
    )

    similarities_per_sections = compare_matched(good_pairs)
    extra_sections_ratio = (len(unmatched_one) + len(unmatched_two)) / (
        len(plan_one.sections) + len(plan_two.sections)
    )
    unmatched_sections_titles = [s.title for s in unmatched_one] + [
        s.title for s in unmatched_two
    ]

    min_similarity = min(
        [pair["content_similarity"] for pair in similarities_per_sections]
    )

    return min_similarity, extra_sections_ratio, unmatched_sections_titles
```

```python
# somewhere to save the overall best and worst similarity scores for easy access
base_min_max = {
    "overall_base_max_sim": float("-inf"),
    "overall_base_max_sim_plans": None,
    "overall_base_max_sim_sections": None,
    "overall_base_max_extra": float("-inf"),
    "overall_base_max_extra_plans": None,
    "overall_base_min_sim": float("+inf"),
    "overall_base_min_sim_plans": None,
    "overall_base_min_extra": float("+inf"),
    "overall_base_min_extra_plans": None,
    "overall_base_min_extra_sections": None,
}


# This will take a while and call openai embeddings once per section title and a bit less than once per section... for my 30 plans it took 5 minutes
def calculate_content_similarity(client_id, min_max):
    # Get all rows for that client, extract plan and markdown
    gen_data = base[base["inputs.client_id"] == client_id]["outputs.gen_data"]
    plans = [
        (
            ActionPlan(**json.loads(data)["structured_action_plan"]),
            json.loads(data)["action_plan"],
        )
        for data in gen_data
    ]
    print(f"found {len(plans)} for client {client_id}")
    min_sim = float("+inf")
    max_extra_sections_ratio = 0
    for i in range(len(plans)):
        for j in range(i + 1, len(plans)):
            (action_plan_i, markdown_i) = plans[i]
            (action_plan_j, markdown_j) = plans[j]

            sim, extra_sections_ratio, extra_sections_titles = get_content_similarity(
                action_plan_i, action_plan_j
            )
            if (sim) > min_max["overall_base_max_sim"]:
                min_max["overall_base_max_sim"] = sim
                min_max["overall_base_max_sim_plans"] = (markdown_i, markdown_j)
            if (sim) < min_max["overall_base_min_sim"]:
                min_max["overall_base_min_sim"] = sim
                min_max["overall_base_min_sim_plans"] = (markdown_i, markdown_j)
            min_sim = min(min_sim, sim)

            if extra_sections_ratio < min_max["overall_base_min_extra"]:
                min_max["overall_base_min_extra"] = extra_sections_ratio
                min_max["overall_base_min_extra_plans"] = (markdown_i, markdown_j)
                min_max["overall_base_min_extra_sections"] = extra_sections_titles
            if extra_sections_ratio > min_max["overall_base_max_extra"]:
                min_max["overall_base_max_extra"] = extra_sections_ratio
                min_max["overall_base_max_extra_plans"] = (markdown_i, markdown_j)
                min_max["overall_base_max_extra_sections"] = extra_sections_titles
            max_extra_sections_ratio = max(
                max_extra_sections_ratio, extra_sections_ratio
            )

    return min_sim, max_extra_sections_ratio


base_diff[["min_content_sim", "extra_sections_ratio"]] = base_diff.index.map(
    lambda i: calculate_content_similarity(i, base_min_max)
).tolist()

print(f"Overall max content: {base_min_max["overall_base_max_sim"]}")
print(f"Overall min content: {base_min_max["overall_base_min_sim"]}")


print(f"Overall max unmatched sections ratio: {base_min_max["overall_base_max_extra"]}")
print(
    f"Overall max unmatched sections ratio: {base_min_max["overall_base_max_extra_sections"]}"
)

print(f"Overall min unmatched sections ratio: {base_min_max["overall_base_min_extra"]}")
print(
    f"Overall max unmatched sections ratio: {base_min_max["overall_base_min_extra_sections"]}"
)
```

```python
# Running  this function will take a while and call openai embeddings once per section title and a bit less than once per section
def calculate_content_similarity_outer(client_id, sim_to_markdown, dataset):
    # Get all rows for that client, show only the markdown column
    gen_data = dataset[dataset["inputs.client_id"] == client_id]["outputs.gen_data"]

    plans_m = [
        (
            ActionPlan(**json.loads(data)["structured_action_plan"]),
            json.loads(data)["action_plan"],
        )
        for data in gen_data
    ]

    print(f"found {len(plans_m)} modified plans for client {client_id}")
    non_null_summary = next(
        (
            x
            for x in dataset[dataset["inputs.client_id"] == client_id]["inputs.summary"]
            if pandas.notna(x)
        ),
        None,
    )
    summary_slice = non_null_summary[50:100]
    print(summary_slice)
    base_id = next(
        (
            x
            for x in base[base["outputs.gen_data"].str.contains(summary_slice)][
                "inputs.client_id"
            ]
            if pandas.notna(x)
        ),
        None,
    )
    if base_id is None:
        base_id = next(
            (
                x
                for x in base[
                    base["inputs.summary"].str.contains(summary_slice, na=False)
                ]["inputs.client_id"]
                if pandas.notna(x)
            ),
            None,
        )
    if base_id is None:
        # manual handling of summaries that I re-generated
        if "Carter" in non_null_summary:
            base_id = next(
                (
                    x
                    for x in base[base["outputs.gen_data"].str.contains("Carter")][
                        "inputs.client_id"
                    ]
                    if pandas.notna(x)
                ),
                None,
            )
        if "Johnson" in non_null_summary:
            base_id = next(
                (
                    x
                    for x in base[base["outputs.gen_data"].str.contains("Johnson")][
                        "inputs.client_id"
                    ]
                    if pandas.notna(x)
                ),
                None,
            )

    print(base_id)
    gen_data = base[base["inputs.client_id"] == base_id]["outputs.gen_data"]
    plans_base = [
        (
            ActionPlan(**json.loads(data)["structured_action_plan"]),
            json.loads(data)["action_plan"],
        )
        for data in gen_data
    ]
    print(f"found {len(plans_base)} base plans")

    min_sim = float("+inf")
    max_ratio = 0
    min_sim_md = None
    for pm in plans_m:
        for pb in plans_base:  # start from i+1 to avoid duplicates
            (action_plan_m, markdown_m) = pm
            (action_plan_base, markdown_base) = pb

            sim, extra_sections_ratio, extra_sections_titles = get_content_similarity(
                action_plan_m, action_plan_base
            )

            if sim < min_sim:
                min_sim_md = (markdown_m, markdown_base)
            min_sim = min(min_sim, sim)
            max_ratio = max(max_ratio, extra_sections_ratio)
    sim_to_markdown.append({"sim": min_sim, "markdowns": min_sim_md})
    return min_sim, max_ratio, base_id
```
