```python
# Imports
from pathlib import Path

from app.core.config import gen_model as model
```

```python
# The case description goes in the clien_system prompt.
# Give as much detail as possible as to who this synthetic client is.

base_path = Path("").parent

# change this !
your_file_path = (
    base_path
    / "experiments"
    / "evaluation"
    / "generate_intakes"
    / "Profiles Input documents"
    / "lex_fisher.txt"
)
background = None
with open(your_file_path) as file:
    background = file.read()

summary_prompt = f"""
--------
Background Information:
{background}
-------

Give a short summary of the person's background. Be direct and precise, but concise.
"""

summary = model.invoke(summary_prompt)
print(summary.content)

needs_and_risk_prompt = f"""
--------
Background Information:
{background}
-------
    Write a mardown report on this person's needs and risk, following this structure:
    ### Needs and Risks Overview
        - **Education:** 
        - **Employment:** 
        - **Financial:** 
        - **Family:** 
        - **Housing:** 
        - **Recreation:** 
        - **Associates:** 
        - **Cognition/Psych:** 
        - **Health:** 

    Keep each section vecry short, one sentence max.
"""

needs_and_risk = model.invoke(needs_and_risk_prompt)
print(needs_and_risk.content)


priority_needs_prompt = f"""
--------
Background Information:
{background}
-------

What are the three to four most urgent domains to adress for this person ?
answer in a very short markdown report in this format:

### Priority Needs
- **[domain]:** short explanation

"""
priority_needs = model.invoke(priority_needs_prompt)
print(priority_needs.content)

full_summary_prompt = f"""
Please correct this text for coherence and conciseness
Do not comment on the text, answer with only your modified version

### Personal Background
{summary.content}

### Needs and Risks Overview
{needs_and_risk.content}

### Priority Needs
{priority_needs.content}

"""

full_summary = model.invoke(full_summary_prompt)
print(full_summary.content)
```

## Save your summary to a json file

```python
from pathlib import Path

base_path = Path("").parent

# change this !
your_file_path = (
    base_path / "experiments" / "evaluation" / "generate_intakes" / "lex_fisher.md"
)


with your_file_path.open("w") as f:
    f.write(full_summary.content)
```
